## **Exploratory Data Analysis on Type of Dwelling**

```python
df_dwelling = df_dwelling.fillna(0)

dwelling_cols = [
    'hdb_1_and_2_room_flats',
    'hdb_3_room_flats',
    'hdb_4_room_flats',
    'hdb_5_room_and_executive_flats',
    'condominiums_and_other_apartments',
    'landed_properties',
    'others'
]

df_dwelling[dwelling_cols] = df_dwelling[dwelling_cols].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
df_dwelling['total_dwelling_units'] = df_dwelling[dwelling_cols].sum(axis=1)

df_dwelling['pct_hdb_1_and_2_room_flats'] = df_dwelling['hdb_1_and_2_room_flats'] / df_dwelling['total_dwelling_units']
df_dwelling['pct_hdb_3_room_flats'] = df_dwelling['hdb_3_room_flats'] / df_dwelling['total_dwelling_units']
df_dwelling['pct_hdb_4_room_flats'] = df_dwelling['hdb_4_room_flats'] / df_dwelling['total_dwelling_units']
df_dwelling['pct_hdb_5_room_and_executive_flats'] = df_dwelling['hdb_5_room_and_executive_flats'] / df_dwelling['total_dwelling_units']
df_dwelling['pct_condominiums_and_other_apartments'] = df_dwelling['condominiums_and_other_apartments'] / df_dwelling['total_dwelling_units']
df_dwelling['pct_landed_properties'] = df_dwelling['landed_properties'] / df_dwelling['total_dwelling_units']
```

```python
df_dwelling_2020 = df_dwelling[df_dwelling['year'] == '2020'].copy() # note that year is a string, not an integer

pct_cols = [col for col in df_dwelling_2020.columns if col.startswith('pct_')]

for col in pct_cols:
    s = pd.to_numeric(df_dwelling_2020[col], errors='coerce')
    s = s[s > 0]   # exclude zeros and NaNs

    if s.empty:
        print(f"\n{col}")
        print("No positive values in 2020")
        continue

    max_idx = s.idxmax()
    min_idx = s.idxmin()

    max_row = df_dwelling_2020.loc[max_idx]
    min_row = df_dwelling_2020.loc[min_idx]

    print(f"\n{col}")
    print(f"Highest: {max_row['planning_area']} ({max_row[col]})")
    print(f"Lowest: {min_row['planning_area']} ({min_row[col]})")
```

## **Exploratory Data Analysis on Transport to School**

```python
df_transport_school = df_transport_school.fillna(0)

transport_school_cols = [
    'bus', 'mrt', 'mrt_bus', 'mrt_car',
    'mrt_other', 'taxi', 'car', 'pvt_chartered_bus', 'lorry_pickup',
    'motorcycle_scooter', 'others', 'no_transport_required',
    'other_combi_mrt_or_bus', 'mrt_lrt_only', 'mrt_lrt_and_bus',
    'other_combi_mrt_lrt_or_bus', 'taxi_pvt_hire_car_only',
    'pvt_chartered_bus_van'
]

df_transport_school[transport_school_cols] = df_transport_school[transport_school_cols].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
df_transport_school['total_transport_school'] = df_transport_school[transport_school_cols].sum(axis=1)

df_transport_school['pct_bus'] = df_transport_school['bus'] / df_transport_school['total_transport_school']
df_transport_school['pct_mrt'] = df_transport_school['mrt'] / df_transport_school['total_transport_school']
df_transport_school['pct_mrt_bus'] = df_transport_school['mrt_bus'] / df_transport_school['total_transport_school']
df_transport_school['pct_car'] = df_transport_school['car'] / df_transport_school['total_transport_school']
df_transport_school['pct_mrt_lrt_only'] = df_transport_school['mrt_lrt_only'] / df_transport_school['total_transport_school']
df_transport_school['pct_mrt_lrt_and_bus'] = df_transport_school['mrt_lrt_and_bus'] / df_transport_school['total_transport_school']
```

```python
df_transport_school_2020 = df_transport_school[df_transport_school['year'] == '2020'].copy() # note that year is a string, not an integer

pct_cols = [col for col in df_transport_school_2020.columns if col.startswith('pct_')]

for col in pct_cols:
    s = pd.to_numeric(df_transport_school_2020[col], errors='coerce')
    s = s[s > 0]   # exclude zeros and NaNs

    if s.empty:
        print(f"\n{col}")
        print("No positive values in 2020")
        continue

    max_idx = s.idxmax()
    min_idx = s.idxmin()

    max_row = df_transport_school_2020.loc[max_idx]
    min_row = df_transport_school_2020.loc[min_idx]

    print(f"\n{col}")
    print(f"Highest: {max_row['planning_area']} ({max_row[col]})")
    print(f"Lowest: {min_row['planning_area']} ({min_row[col]})")
```
