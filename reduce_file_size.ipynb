```python
import os
import pandas as pd

CSV_PATH = 'data/transform_resale_flat_price.csv'

df = pd.read_csv(CSV_PATH)
orig_size_mb = os.path.getsize(CSV_PATH) / (1024 ** 2)
print(f"Original: {len(df)} rows, {len(df.columns)} columns, {orig_size_mb:.1f} MB")
print("\nAll columns:")
for c in df.columns:
    print(f"  {c}")
```

```python
NEEDED_COLS = [
    'latitude', 'longitude',
    'town', 'max_floor_lvl', 'total_dwelling_units',
    'dist_to_nearest_mrt_m', 'n_mrt_within_1km',
    'dist_to_nearest_bus_stop_m', 'n_bus_stop_within_1km',
    'dist_to_food_m', 'n_food_within_1km',
    'dist_to_supermarket_m', 'n_supermarket_within_1km',
    'month_and_year', 'resale_price',
]


csv_cols   = set(df.columns.str.lower())
needed_set = set(c.lower() for c in NEEDED_COLS)

dropping = sorted(csv_cols - needed_set)
missing  = needed_set - csv_cols

print(f"Columns in CSV      : {len(csv_cols)}")
print(f"Columns needed      : {len(needed_set)}")
print(f"Columns to drop     : {len(dropping)}")
print(f"Needed cols missing : {missing or 'none'}")
print("\nDropping:")
for c in dropping:
    print(f"  {c}")
```

```python
keep_cols = [c for c in df.columns if c.lower() in needed_set]

df_reduced = df[keep_cols].copy()
df_reduced.to_csv(CSV_PATH, index=False)

new_size_mb = os.path.getsize(CSV_PATH) / (1024 ** 2)
saved_mb    = orig_size_mb - new_size_mb

print(f"Reduced : {len(df_reduced)} rows, {len(df_reduced.columns)} columns, {new_size_mb:.1f} MB")
print(f"Saved   : {saved_mb:.1f} MB  ({100 * saved_mb / orig_size_mb:.1f}% reduction)")
print(f"\nKept columns: {list(df_reduced.columns)}")
```
